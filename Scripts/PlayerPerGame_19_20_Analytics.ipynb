```python
def stats(stats_list):
    global df_player_19_20
    
    for x in stats_list:
        df_player_19_20 = df_player_19_20.sort_values(by=x, ascending=False)
        fig, ax = plt.subplots(figsize=(20,6))
        ax = sns.barplot(x=df_player_19_20.iloc[0:30, 1], y=df_player_19_20[x], data=df_player_19_20, 
                         dodge=False)
        ax.set_xlabel(xlabel='Players')
        ax.set_ylabel(ylabel=str(x))
        
        #rects=ax.patches
        #labels=['{:04.2f}'.format(i) for i in df_player_19_20[x]]
        
        #for rect, label in zip(rects, labels):
            #height = rect.get_height()
            #ax.text(rect.get_x() + rect.get_width() / 2, height + 0.02, label,
                            #ha='center', va='bottom')
        
        plt.xticks(rotation=90)
        plt.show()
```
